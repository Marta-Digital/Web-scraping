```python
# Creamos la clase JsonWriterPipeline, que se encargará de convertir todos los elementos conseguidos a un archivo JSON, donde 
# cada linea es un elemento.

class JsonWriterPipeline(object):
    def open_spider(self, spider):
        self.file = open('resultados.json', 'w')

    def close_spider(self, spider):
        self.file.close()

    def process_item(self, item, spider):
        line = json.dumps(ItemAdapter(item).asdict()) + "\n"
        self.file.write(line)
```

```python
# Definimos un nuevo Item al que llamaremos Producto

class Producto(scrapy.Item):
    articulo = scrapy.Field()
    precio = scrapy.Field()
    descripcion = scrapy.Field()
```

```python
# Creamos nuestra spider con las URLs de 10 prendas que más me gustan de la nueva colección de Zara Fall/Winter'21

class ZaraSpider(scrapy.Spider):
    name = "zara"
    allowed_domains = ['zara.com'] # Dominios permitidos 
    start_urls = ['https://www.zara.com/es/es/blazer-con-lino-rayas-p07791667.html?v1=114779319&v2=1882227',
                  'https://www.zara.com/es/es/velvet-pigment-lip-gloss-p24130318.html?v1=116022356&v2=1884173',
                  'https://www.zara.com/es/es/vestido-mini-estampado-p07521261.html?v1=108972051&v2=1882775',
                  'https://www.zara.com/es/es/vestido-mini-frunces-p00085805.html?v1=119884478&v2=1882775',
                  'https://www.zara.com/es/es/pack-pinzas-con-brillo-p04319001.html?v1=111229420&v2=1883491',
                  'https://www.zara.com/es/es/zueco-tachas-p13502810.html?v1=109007025&v2=1883006',
                  'https://www.zara.com/es/es/camiseta-oversize-rayas-p00858072.html?v1=109708972',
                  'https://www.zara.com/es/es/jeans-z1975-regular-fit-p05862151.html?v1=108928722&v2=1882892',
                  'https://www.zara.com/es/es/jeans-z1975-mom-fit-p05862097.html?v1=108944929&v2=1882892',
                  'https://www.zara.com/es/es/cuerpo-lencero-satinado-p07385415.html?v1=114966338&v2=1882956']
    
    custom_settings = {'LOG_LEVEL': logging.WARNING,
                       'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 1}, 
                       'FEED_FORMAT':'json',                                 
                       'FEED_URI': 'resultados.json',
                       'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}   
    
    def parse(self, response):
        loader = ItemLoader(item=Producto())
        loader.add_value('articulo', [response.xpath('//h1[@class="product-detail-info__name"]/text()').extract_first()])
        loader.add_value('precio', [response.xpath('//span[@class="price__amount-current"]/text()').extract_first()])
        loader.add_value('descripcion', [response.xpath('//*[@id="main"]/article/div[2]/div[2]/div[1]/p/text()[2]').extract_first()])
        
        
        yield loader.load_item()
```

```python
# Ahora ya podemos ejecutar nuestro spider ZaraSpider

process = CrawlerProcess(get_project_settings())
process.crawl(ZaraSpider)
process.start()
```
